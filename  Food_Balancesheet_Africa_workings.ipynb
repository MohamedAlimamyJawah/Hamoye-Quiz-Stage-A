## QUESTION: 
Select columns â€˜Y2017â€™ and â€˜Areaâ€™, Perform a groupby operation on â€˜Areaâ€™.  Which of these Areas had the 7th lowest sum in 2017?

```python
# Select 'Y2017' and 'Area', group by 'Area', and sum 'Y2017'
grouped = Food_Balancesheet_Africa[['Area', 'Y2017']].groupby('Area').sum()

# Sort the sums in ascending order and select the 7th entry
seventh_lowest = grouped.sort_values('Y2017').iloc[6]

print(seventh_lowest)
```

## QUESTION :
Perform a groupby operation on â€˜Elementâ€™.  What year has the highest sum of Stock Variation?

```python
# Stock variation

Food_Balancesheet_Africa['Stock Variation'] =  Food_Balancesheet_Africa[['Y2014', 'Y2015', 'Y2016', 'Y2017']].diff(axis=1).abs().sum(axis=1)
grouped =  Food_Balancesheet_Africa.groupby('Element')['Stock Variation'].sum()
```

```python
# Find the year with the highest sum of stock variation
max_year =  Food_Balancesheet_Africa.loc[:, 'Y2014':'Y2017'].sum().idxmax()

print(max_year)
```
